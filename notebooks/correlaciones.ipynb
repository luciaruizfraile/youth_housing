```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Carga de los CSVs
df_precios = pd.read_csv("../data/clean/precios_sql.csv")
df_salario_ccaa = pd.read_csv("../data/clean/salario_ccaa_sql.csv")
df_salario_edad = pd.read_csv("../data/clean/salario_edad_sql.csv")
df_interes = pd.read_csv("../data/clean/interes_fijo_sql.csv")
df_superficie = pd.read_csv("../data/clean/superficie_extendida.csv")
df_ipv = pd.read_csv("../data/clean/ipv.csv")

# Funcin para crear la matriz de correlacin y graficar heatmap
def matriz_correlacion(df, titulo='Matriz de Correlacin'):
    corr = df.corr(numeric_only=True)
    plt.figure(figsize=(10, 8))
    sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
    plt.title(titulo)
    plt.show()

# Mostrar heatmaps para cada dataset
matriz_correlacion(df_precios, "Correlacin - Precios")
matriz_correlacion(df_salario_ccaa, "Correlacin - Salario por CCAA")
matriz_correlacion(df_salario_edad, "Correlacin - Salario por Edad")
matriz_correlacion(df_interes, "Correlacin - Inters")
matriz_correlacion(df_superficie, "Correlacin - Superficie")
matriz_correlacion(df_ipv, "Correlacin - IPV")
```
